# Importing necessary libraries

```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from collections import Counter
```

# Generating Synthetic Dataset

```python
X,y = make_blobs(n_samples = 1000_000,
                 centers = 5,
                 n_features = 2,
                 random_state = 42)
```

```python
X.shape, y.shape
```

# Visualization

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', edgecolor='k')
plt.show()
```

# Model Class

```python
class KNN:
    def __init__(self, X, y, k=7):
        self.X = X  # Features
        self.y = y  # Labels
        self.k = k  # Number of neighbours
    
    def return_class(self, x0):
        diff = self.X - x0
        distances = np.sqrt(np.sum(diff ** 2, axis=1))
        
        k_indices = np.argsort(distances)[:self.k]
    
        k_labels = self.y[k_indices]
        
        freq_map = Counter(k_labels)
        maj_label = freq_map.most_common(1)[0][0]
        
        return int(maj_label)
```

# Implementation

```python
knn = KNN(X,y,k=5)
q = np.array([-1.7,8.19])

knn.return_class(q)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', edgecolor='k')

q = np.array([-1.7, 8.19])
plt.scatter(q[0], q[1], c='red', marker='X', s=200, edgecolor='black', label='Query Point')

plt.legend()
plt.show()
```
