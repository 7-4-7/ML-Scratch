# Linear Regression on Diabetes Dataset

```python
import numpy as np
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
import seaborn as sns
```

# Getting the data

```python
data_matrix = load_diabetes()
```

```python
X = data_matrix['data']
y = data_matrix['target']
```

# Checking out linear dependency

```python
gram_matrix = X.T @ X

sns.heatmap(gram_matrix)
plt.show() # Looks good 
```

# Model building

```python
class LR:
    def __init__(self, X, y):
        self.X = self.modify_x(X)
        self.y = y
        self.beta = None
        
    def modify_x(self, X):
        return np.hstack([np.ones((X.shape[0], 1)), X])
        
    def fit(self):
        gram_matrix = self.X.T @ self.X
        self.beta = np.linalg.inv(gram_matrix) @ self.X.T @ self.y
    
    def predict(self, X_0):
        mod_x = self.modify_x(X_0)
        return mod_x @ self.beta
```

```python
model = LR(X,y)
model.fit()
```
